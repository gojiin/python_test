### 이상 데이터 정제하기 #2 : 함수 사용

```python
def refiningmelong(df) :
    if df['PRODUCTAGE'] < 1 :
        return 1
    elif df['PRODUCTAGE'] < 2 :
        return 2
    else :
        return 5
    
customerData["PRODUCTAGE_New2"] = customerData.apply(refiningmelong, axis=1)
```

### 데이터 요약분석하기 : groupby 이용

```python
groupKey = ["CUSTTYPE"]
```

```python
# 평균 및 표준편차 구하기
groupData = customerData.\
groupby(groupKey).\
agg(['mean','std'])
groupData
```

```python
groupData = groupData.reset_index()  ##groupby 함수는 종료 후 reset_index()를 통해 인덱스를 초기화시켜야함
groupData.head()
```
