## 데이터 불러오기 및 데이터 조작하기

```python
## 데이터 불러오기

import pandas as pd

customerData2 = pd.read_csv("./dataset/kopo_product_volume.csv")
```

```python
## 조건 설정하여 데이터 추출하기
answer = customerData2[(customerData2.PRODUCTGROUP=="ST0002") & (customerData2.VOLUME>300000)]
## 대소문자 잘 가려줘야함!, 문자열은 큰따옴표로 감싸줘야함!
answer
```

```python
## query를 이용하여 조건 설정 간단하게 한 후 추출하기
answer2 = customerData2.query('PRODUCTGROUP == "ST0002" & VOLUME>300000')
answer2
```
